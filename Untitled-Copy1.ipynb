```python
def remove_useless(medicine_name):
    last_left = medicine_name.rfind('(')
    last_right = medicine_name.rfind(')')
    if last_left!=-1 and last_right!=-1 and last_left<last_right:
        return medicine_name[:last_left]
    else:
        return medicine_name

remove_useless('凉血解毒颗粒(华威)')
remove_useless('夏桑菊颗粒(九后爽)')
```
